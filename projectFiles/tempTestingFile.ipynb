```python
import re 
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def ytUrlId(url: str) -> str:
    # using the regex 
    pattern = r"(?:v=|\/)([0-9A-Za-z_-]{11}).*"
    match = re.search(pattern, url)

    return match.group(1) if match else None
```

```python
# uses the ytTranscript Api to fetch the transcripts of a yt video 
def transcipt(ytID: str , lang: str = "en") -> str:
    ytAPI = YouTubeTranscriptApi()
    res = ytAPI.fetch(ytID , languages = [lang])

    return res
```

```python
res = ytUrlId("https://www.youtube.com/watch?v=TUKk3sptrt0")
```

```python
out = transcipt(res,'hi')
```
